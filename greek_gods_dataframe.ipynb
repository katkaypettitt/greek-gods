```python
import greek_gods_lists
import pandas as pd

pd.set_option('display.max_rows', None)
```

```python
names_english = greek_gods_lists.name_english
names_greek = greek_gods_lists.name_greek
main_types = greek_gods_lists.t
sub_types = greek_gods_lists.t_sub
descriptions = greek_gods_lists.description
```

```python
combined = {'name-english': names_english,'name-greek': names_greek, 
            'main-type': main_types, 'sub-type': sub_types,
            'description': descriptions}
```

```python
df = pd.DataFrame(combined)
df = df.astype(str)
```

```python
df
```

```python
df['name-english'] = df['name-english'].apply(lambda x: x.replace("['",'').replace("']",'')) 
df['name-greek'] = df['name-greek'].apply(lambda x: x.replace("['",'').replace("']",'')) 
df['description'] = df['description'].apply(lambda x: x.replace("[",'').replace("]",'')
                                            .replace('"', '').replace("'", '')) 
```

```python
df.drop(df[df['name-english'] == '[]'].index, inplace=True)
df = df.reset_index(drop=True)
```

```python
df['description'] = df['description'].str.lower()
df['description'] = df['description'].str.split('.').str[0]
```

```python
df['description'] = df['description'].map(lambda x: x[3:] if x.startswith('the ') else x)
df['description'] = df['description'].map(lambda x: x[2:] if x.startswith('an ') else x)
df['description'] = df['description'].map(lambda x: x[1:] if x.startswith('a ') else x)
```
